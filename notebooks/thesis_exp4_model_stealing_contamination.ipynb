# Experiment 4: Model Stealing and Contamination Detection

Goal: test whether Adaptive ISQED can distinguish exact peer clones from fine-tuned near-duplicate targets.

We reuse the Exp3 prompt-intervention audit pipeline.

Targets:
- Perfect Redundancy (Clone)
- Fine-tuned Variant
- Architectural Divergence target

Peers are fixed within each task.

Methods:
- Random
- JointPairUCB

Metrics:
- pointwise family holdout residual
- best-so-far residual under matched query cost

The clone target should produce near-zero residuals.
The fine-tuned target should remain close to peers on many contexts but expose high-residual contexts in the tail.
